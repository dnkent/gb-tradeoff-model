## - *Core Question*: How can states limit the costs of anarchy, understood as mutually undesirable but necessary military expenditures?

![image](arms_races.png)

## *- Argument:* The costs of anarchy are primarily a function of domestic politics.

###    - International structure servers as a permissive factor: creating environments where combinations of domestic configurations can create more or less severe military competitions.

## - *Existing arguments*:

### Generaly a derivation of the spiral model (Jervis, 1978) and hinge upon two variables (i.e. Fearon, 2018):
### 1. **offense-defense balance**
### 2. **distinguishability of intentions** 

## Shortcomings
### -  Intentions are rarely clear and can always change in the future. (Copeland, 2000)
### - Also there are few, if any, true indices of current intentions. (Jervis, 1989)
### -  The offense-defense balance is incredibly difficult to operationalize for scholars and statesmen alike. (Biddle, 2001; Glaser and Kaufmann, 1998)
### -  Defensive capabilities can generally be converted to offense relatively easily. (Conversations with Gelpi)

## Payoff: Based on theory, we should expect spirals and intense military competition abound. 
## But, even for enduring rivalries, said costs are clearly limited.

![rivals](rivals.png)

## Why? State-level variables are a serious limitation.
 

## An ABM of Arming
**Setup**
- 14x14 grid with one agent per cell, each agent is a state. 
- State wealth is drawn from a pareto distribution with a scale parameter of 2.
- State domestic needs are a total wealth multipled by a constant percent.
- Growth is drawn from a truncated normal distribution (so there is no negative growth) with a mean of 3. 
- Starting arms are a randomly drawn percent of GDP.
- Grid wraps as an approximation for world. 
- Agents are all activated once per step.

## Load the model and a packages

```python
from model2 import ArmsModel
from arming_econ_same import EconMod

%matplotlib inline
import matplotlib.pyplot as plt

import random
import numpy as np
import pandas as pd
```

## Minimal Guns-Butter Tradeoff

```python
height = 14
width = 14
density = 1
domestic_min = 0.01
domestic_max = 0.1
random.seed(123)
```

```python
model = EconMod(height, width, density,domestic_min, domestic_max)
```

## Starting Conditions

### Arms

```python
agent_arms = [a.arms for a in model.schedule.agents]
```

```python
plt.figure(figsize = (8, 8))
plt.hist(agent_arms)
# plt.savefig('start_hist.png', bbox_inches='tight') ## Save image
plt.show() ## Don't show array output
```

```python
agent_arms = np.zeros((model.grid.width, model.grid.height))
for a in model.schedule.agents:
    X = a.pos[0]
    Y = a.pos[1]
    arms = a.arms
    agent_arms[X][Y] = arms
```

```python
plt.figure(figsize = (13, 9))
plt.imshow(agent_arms, interpolation='nearest', aspect = 'auto', cmap="Blues")
plt.colorbar()
#plt.savefig('arms_start.pdf', bbox_inches='tight') ## Save image
plt.show()
```

```python
agent_burden = np.zeros((model.grid.width, model.grid.height))
for a in model.schedule.agents:
    X = a.pos[0]
    Y = a.pos[1]
    burden = a.mil_burden
    agent_burden[X][Y] = burden
```

## Secondary Implication: Spiral vs. Deterrence Models and War

### - Here, we get one quasi-arms race -- one state can't truly keep up. Often there are no arms races and just regional hegemons.
### - **Interestingly, the model potentially synthesizes seemingly competing models of security: the spiral and deterrence models** (Braumoeller, 2008)
### - The arming state has a strategic advantage over neighbors, but also induces an undesirable spiral with the other outlier state. Ideally both would stick with the pack spending-wise. All other states succesfully deter each other at low-levels of spending, but they also fail to deter against the leading powers.
